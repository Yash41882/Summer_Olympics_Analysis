# Summer Olympics Data Analysis Assignment


```python
# Importing the Necessary Libraries/Modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the Summer Olympics.csv File

df = pd.read_csv("Summer Olympics.csv")
```

```python
df.head()
```

## 1. In how many cities Summer Olympics is held so far?

```python
record = df['City'].unique()
print("Total Number of Cities = ",len(record))
print("Name of Cities = ", record)
```

```python
len(df['City'].unique())
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
record1 = []
gold = df[df['Medal']=='Gold']
for sport in gold['Sport'].unique():
    record1.append([sport , len(gold[gold['Sport']  == sport])])

pd.DataFrame(record1,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

```python
df_pr = df[df['Medal'] == 'Gold']

df_pr.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (5,5))
```

```python
df_pr.groupby('Sport').count()['Medal'].sort_values(ascending = False).head()
```
